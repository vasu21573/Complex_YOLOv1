```python
import torch
import matplotlib.pyplot as plt
```

```python
train_loss = torch.load('/content/drive/MyDrive/VOC2007/Train/complex_loss.pt', map_location=torch.device('cpu'))
```

```python
com_train_loss = []
sum = 0
for i in range(len(train_loss)):
  sum += train_loss[i].item()
  if (i + 1) % 3 == 0:
    com_train_loss.append(sum)
    sum = 0
```

```python
train_loss = torch.load('/content/drive/MyDrive/VOC2007/Train/real_loss.pt', map_location=torch.device('cpu'))
```

```python
norm_train_loss = []
sum = 0
for i in range(len(train_loss)):
  sum += train_loss[i].item()
  if (i + 1) % 3 == 0:
    norm_train_loss.append(sum)
    sum = 0
```

```python
val_loss = torch.load('/content/drive/MyDrive/VOC2007/Validation/complex_loss.pt', map_location=torch.device('cpu'))
```

```python
com_val_loss = []
for i in range(len(val_loss)):
  com_val_loss.append(val_loss[i].item())
```

```python
val_loss = torch.load('/content/drive/MyDrive/VOC2007/Validation/real_loss.pt', map_location=torch.device('cpu'))
```

```python
norm_val_loss = []
for i in range(len(val_loss)):
  norm_val_loss.append(val_loss[i].item())
```
